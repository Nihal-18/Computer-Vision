Perform gaussian filtering operation in an image

```python
import cv2
import numpy as np
```

```python
from google.colab.patches import cv2_imshow
```

```python
# reading image using cv2.imread()
in_img= cv2.imread('/content/drive/MyDrive/Unknown-2')
```

```python
cv2_imshow(in_img)
```

```python
# Applying Gaussian Filter

op_img = cv2.GaussianBlur(in_img, (5,5), cv2.BORDER_DEFAULT)
```

```python
# Displaying image after Gaussian Filtering was applied

cv2_imshow(op_img)
```

```python
# Applying Gaussian Filter by increasing the kernel size 

# NOTE: Only odd kernel sizes are allowed 

op_img_1 = cv2.GaussianBlur(in_img, (11,11), cv2.BORDER_DEFAULT)
```

```python
# Displaying image after Gaussian Filtering was applied after increasing the kernel size

cv2_imshow(op_img_1)
```
