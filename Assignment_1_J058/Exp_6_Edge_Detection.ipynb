Perform image edge detection using sobel and canny filtering operations

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab.patches import cv2_imshow
```

```python
# reading image using cv2.imread()
img0= cv2.imread('/content/drive/MyDrive/Unknown-2')
```

```python
cv2_imshow(img0)
```

```python
# converting to gray scale
gray = cv2.cvtColor(img0, cv2.COLOR_BGR2GRAY)

# remove noise
img = cv2.GaussianBlur(gray,(3,3),0)
```

# Sobel filter

```python
# Appyling sobel operator over x-axis of the gray image

sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)  # x


cv2_imshow(sobelx)
```

```python
# Appyling sobel operator over x-axis of the gray image

sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)  # y

cv2_imshow(sobely)
```

# Canny Filter

```python
# Applying canny filter

# 1st argument: source of input image

# 2nd argument: minVal threshold

# 3rd argument: maxVal threshold

# 4th argument: kernel size of sobel filter (default= 3)

can_img= cv2.Canny(gray, 100, 200, 5)

plt.subplot(121),plt.imshow(gray,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(can_img,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
```
