# Exp 8: 
Perform opening and closing morphological operations in an image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab.patches import cv2_imshow
```

```python
# reading image using cv2.imread()
img0= cv2.imread('/content/drive/MyDrive/Unknown-2')
```

```python
cv2_imshow(img0)
```

```python
# converting to gray scale
gray = cv2.cvtColor(img0, cv2.COLOR_BGR2GRAY)

# remove noise
img = cv2.GaussianBlur(gray,(3,3),0)
```

```python
# Applying opening morphological operation on the image

kernel= np.ones(shape= (5,5))

img_open = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)

cv2_imshow(img_open)
```

```python
# Applying closing morphological operation on the image

img_close = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

cv2_imshow(img_close)
```
