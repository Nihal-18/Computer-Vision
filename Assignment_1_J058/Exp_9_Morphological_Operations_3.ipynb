# Exp 9: 
Perform morphological gradient, top hat and black hat morphological operations in an image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab.patches import cv2_imshow
```

```python
# reading image using cv2.imread()
img0= cv2.imread('/content/drive/MyDrive/Unknown-2')
```

```python
cv2_imshow(img0)
```

```python
# converting to gray scale
gray = cv2.cvtColor(img0, cv2.COLOR_BGR2GRAY)

# remove noise
img = cv2.GaussianBlur(gray,(3,3),0)
```

```python
# Applying gradient morphological operation on the image

kernel= np.ones(shape= (5,5))

img_grad = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)

cv2_imshow(img_grad)
```

```python
# Applying top hat morphological operation on the image

img_top = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)

cv2_imshow(img_top)
```

```python
# Applying black hat morphological operation on the image

img_black = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernel)

cv2_imshow(img_black)
```
