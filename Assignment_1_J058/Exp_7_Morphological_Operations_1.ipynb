# EXP 7: 
Perform erosion and dilation morphological operations in an image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab.patches import cv2_imshow
```

```python
# reading image using cv2.imread()
img0= cv2.imread('/content/drive/MyDrive/Unknown-2')
```

```python
cv2_imshow(img0)
```

```python
# converting to gray scale
gray = cv2.cvtColor(img0, cv2.COLOR_BGR2GRAY)

# remove noise
img = cv2.GaussianBlur(gray,(3,3),0)
```

```python
# Appling erosion 

kernel = np.ones((5,5),np.uint8)

img_erode = cv2.erode(img, kernel, iterations = 1)

cv2_imshow(img_erode)
```

```python
# Applying Dilation

img_dil = cv2.dilate(img,kernel,iterations = 1)

cv2_imshow(img_dil)
```
