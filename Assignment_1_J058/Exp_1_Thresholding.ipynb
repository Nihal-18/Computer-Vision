#Exp 1: 
Perform thresholding operation in an image and evaluate its histogram using histogram equalisation

```python
import cv2
import numpy as np
```

```python
from google.colab.patches import cv2_imshow
```

```python
# reading image using cv2.imread()
img= cv2.imread('/content/drive/MyDrive/Unknown-2')
```

```python
cv2_imshow(img)
```

```python
# for carryin gout the thresholding operation first
# we convert it to grayscale 

img1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2_imshow(img1)
```

```python
# applying different thresholding 
# techniques on the input image
# all pixels value above 120 will 
# be set to 255

ret, thresh1 = cv2.threshold(img1, 120, 255, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(img1, 120, 255, cv2.THRESH_BINARY_INV)
ret, thresh3 = cv2.threshold(img1, 120, 255, cv2.THRESH_TRUNC)
ret, thresh4 = cv2.threshold(img1, 120, 255, cv2.THRESH_TOZERO)
ret, thresh5 = cv2.threshold(img1, 120, 255, cv2.THRESH_TOZERO_INV)
```

```python
# the window showing output images
# with the corresponding thresholding 
# techniques applied to the input images

print('Binary Threshold')

equ1 = cv2.equalizeHist(thresh1)

res1 = np.hstack((thresh1, equ1))

cv2_imshow(res1)
```

```python
print('Binary Threshold Inverted' )

equ2 = cv2.equalizeHist(thresh2)

res2 = np.hstack((thresh2, equ2))

cv2_imshow(res2)
```

```python
print('Truncated Threshold')

equ3 = cv2.equalizeHist(thresh3)

res3 = np.hstack((thresh3, equ3))

cv2_imshow(res3)
```

```python
print('Set to 0')

equ4 = cv2.equalizeHist(thresh4)

res4 = np.hstack((thresh4, equ4))

cv2_imshow(res4)
```

```python
print('Set to 0 Inverted')

equ5 = cv2.equalizeHist(thresh5)

res5 = np.hstack((thresh5, equ5))

cv2_imshow(res5)
```
